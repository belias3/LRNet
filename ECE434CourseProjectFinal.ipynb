Find Face Location - Sample Fake

```python
from google.colab.patches import cv2_imshow
import cv2
import dlib

#load detector
detector = dlib.get_frontal_face_detector()

#read image
vidcap = cv2.VideoCapture('/content/gdrive/MyDrive/434CourseProj/LRNet-main/demo/input/sample_fake.mp4')
success,image = vidcap.read()

# get vcap property 
width  = vidcap.get(3)  # float `width`
height = vidcap.get(4)  # float `height`

#convert image to grayscale
gray = cv2.cvtColor(src=image, code=cv2.COLOR_BGR2GRAY)

#Use detector to find landmarks
faces = detector(gray)

for face in faces:
    x1 = int(face.left()-0.018*width)
    y1 = int(face.top()-0.1157*height)
    x2 = int(face.right()+0.018*width)
    y2 = int(face.bottom()+0.046*height)
    #draw rectangle
    cv2.rectangle(img = image, pt1=(x1,y1), pt2=(x2,y2), color=(0, 255, 0), thickness=4)
    difw = x2-x1
    difh = y2-y1
    
#show image
cv2_imshow(image)

cv2.destroyAllWindows()
```
